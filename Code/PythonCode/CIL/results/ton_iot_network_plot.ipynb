```python
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
ton_iot_network_finetune = json.load(open("./ton-iot-network_finetune_ton_iot_network_ann_128_Accuracy_curve.json"))
ton_iot_network_finetune = [i[2] for i in ton_iot_network_finetune]
ton_iot_network_der = json.load(open("./ton-iot-network_der_ton_iot_network_ann_128_Accuracy_curve.json"))
ton_iot_network_der = [i[2] for i in ton_iot_network_der]
ton_iot_network_memo = json.load(open("./ton-iot-network_memo_ton_iot_network_memo_ann_128_Accuracy_curve.json"))
ton_iot_network_memo = [i[2] for i in ton_iot_network_memo]
ton_iot_network_lwf = json.load(open("./ton-iot-network_lwf_ton_iot_network_ann_128_Accuracy_curve.json"))
ton_iot_network_lwf = [i[2] for i in ton_iot_network_lwf]
ton_iot_network_icarl = json.load(open("./ton-iot-network_icarl_ton_iot_network_ann_128_Accuracy_curve.json"))
ton_iot_network_icarl = [i[2] for i in ton_iot_network_icarl]
print("Finetune")
print("Avg: ", np.mean(ton_iot_network_finetune))
print("Last:", ton_iot_network_finetune[-1])
print("DER")
print("Avg: ", np.mean(ton_iot_network_der))
print("Last:", ton_iot_network_der[-1])
print("MeMO")
print("Avg: ", np.mean(ton_iot_network_memo))
print("Last:", ton_iot_network_memo[-1])
print("Lwf")
print("Avg: ", np.mean(ton_iot_network_lwf))
print("Last:", ton_iot_network_lwf[-1])
print("Icarl")
print("Avg: ", np.mean(ton_iot_network_icarl))
print("Last:", ton_iot_network_icarl[-1])
```

```python
plt.plot(ton_iot_network_der, label="der")
plt.plot(ton_iot_network_finetune, label="finetune")
plt.plot(ton_iot_network_memo, label="memo")
plt.plot(ton_iot_network_lwf, label="lwf")
plt.plot(ton_iot_network_icarl, label="icarl")
plt.legend()
plt.xlabel("Task")
plt.ylabel("Accuracy")
plt.savefig("ton_iot_network_accy.png")
```
