```python
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
kdd_finetune = json.load(open("./kdd99_finetune_kdd_ann_128_Accuracy_curve.json"))
kdd_finetune = [i[2] for i in kdd_finetune]
kdd_der = json.load(open("./kdd99_der_kdd_ann_128_Accuracy_curve.json"))
kdd_der = [i[2] for i in kdd_der]
kdd_memo = json.load(open("./kdd99_memo_kdd_memo_ann_128_Accuracy_curve.json"))
kdd_memo = [i[2] for i in kdd_memo]
kdd_lwf = json.load(open("./kdd99_lwf_kdd_ann_128_Accuracy_curve.json"))
kdd_lwf = [i[2] for i in kdd_lwf]
kdd_icarl = json.load(open("./kdd99_icarl_kdd_ann_128_Accuracy_curve.json"))
kdd_icarl = [i[2] for i in kdd_icarl]
print("Finetune")
print("Avg: ", np.mean(kdd_finetune))
print("Last:", kdd_finetune[-1])
print("DER")
print("Avg: ", np.mean(kdd_der))
print("Last:", kdd_der[-1])
print("MeMO")
print("Avg: ", np.mean(kdd_memo))
print("Last:", kdd_memo[-1])
print("Lwf")
print("Avg: ", np.mean(kdd_lwf))
print("Last:", kdd_lwf[-1])
print("Icarl")
print("Avg: ", np.mean(kdd_icarl))
print("Last:", kdd_icarl[-1])
```

```python
plt.plot(kdd_der, label="der")
plt.plot(kdd_finetune, label="finetune")
plt.plot(kdd_memo, label="memo")
plt.plot(kdd_lwf, label="lwf")
plt.plot(kdd_icarl, label="icarl")
plt.legend()
plt.xlabel("Task")
plt.ylabel("Accuracy")
plt.savefig("kdd_accy.png")
```
