```python
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
cic_ids_2017_finetune = json.load(open("./cic-ids-2017_finetune_cic_ids_ann_128_Accuracy_curve.json"))
cic_ids_2017_finetune = [i[2] for i in cic_ids_2017_finetune]
cic_ids_2017_der = json.load(open("./cic-ids-2017_der_cic_ids_ann_128_Accuracy_curve.json"))
cic_ids_2017_der = [i[2] for i in cic_ids_2017_der]
cic_ids_2017_memo = json.load(open("./cic-ids-2017_memo_cic_ids_memo_ann_128_Accuracy_curve.json"))
cic_ids_2017_memo = [i[2] for i in cic_ids_2017_memo]
cic_ids_2017_lwf = json.load(open("./cic-ids-2017_lwf_cic_ids_ann_128_Accuracy_curve.json"))
cic_ids_2017_lwf = [i[2] for i in cic_ids_2017_lwf]
cic_ids_2017_icarl = json.load(open("./cic-ids-2017_icarl_cic_ids_ann_128_Accuracy_curve.json"))
cic_ids_2017_icarl = [i[2] for i in cic_ids_2017_icarl]
print("Finetune")
print("Avg: ", np.mean(cic_ids_2017_finetune))
print("Last:", cic_ids_2017_finetune[-1])
print("DER")
print("Avg: ", np.mean(cic_ids_2017_der))
print("Last:", cic_ids_2017_der[-1])
print("MeMO")
print("Avg: ", np.mean(cic_ids_2017_memo))
print("Last:", cic_ids_2017_memo[-1])
print("Lwf")
print("Avg: ", np.mean(cic_ids_2017_lwf))
print("Last:", cic_ids_2017_lwf[-1])
print("Icarl")
print("Avg: ", np.mean(cic_ids_2017_icarl))
print("Last:", cic_ids_2017_icarl[-1])
```

```python
plt.plot(cic_ids_2017_der, label="der")
plt.plot(cic_ids_2017_finetune, label="finetune")
plt.plot(cic_ids_2017_memo, label="memo")
plt.plot(cic_ids_2017_lwf, label="lwf")
plt.plot(cic_ids_2017_icarl, label="icarl")
plt.legend()
plt.xlabel("Task")
plt.ylabel("Accuracy")
plt.savefig("cic_ids_accy.png")
```
