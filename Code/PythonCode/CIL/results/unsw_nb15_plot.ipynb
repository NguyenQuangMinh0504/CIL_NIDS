```python
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
unsw_nb15_finetune = json.load(open("./unsw-nb15_finetune_unsw_nb15_ann_128_Accuracy_curve.json"))
unsw_nb15_finetune = [i[2] for i in unsw_nb15_finetune]
unsw_nb15_der = json.load(open("./unsw-nb15_der_unsw_nb15_ann_128_Accuracy_curve.json"))
unsw_nb15_der = [i[2] for i in unsw_nb15_der]
unsw_nb15_memo = json.load(open("./unsw-nb15_memo_unsw_nb15_memo_ann_128_Accuracy_curve.json"))
unsw_nb15_memo = [i[2] for i in unsw_nb15_memo]
unsw_nb15_lwf = json.load(open("./unsw-nb15_lwf_unsw_nb15_ann_128_Accuracy_curve.json"))
unsw_nb15_lwf = [i[2] for i in unsw_nb15_lwf]
unsw_nb15_icarl = json.load(open("./unsw-nb15_icarl_unsw_nb15_ann_128_Accuracy_curve.json"))
unsw_nb15_icarl = [i[2] for i in unsw_nb15_icarl]
print("Finetune")
print("Avg: ", np.mean(unsw_nb15_finetune))
print("Last:", unsw_nb15_finetune[-1])
print("DER")
print("Avg: ", np.mean(unsw_nb15_der))
print("Last:", unsw_nb15_der[-1])
print("MeMO")
print("Avg: ", np.mean(unsw_nb15_memo))
print("Last:", unsw_nb15_memo[-1])
print("Lwf")
print("Avg: ", np.mean(unsw_nb15_lwf))
print("Last:", unsw_nb15_lwf[-1])
print("Icarl")
print("Avg: ", np.mean(unsw_nb15_icarl))
print("Last:", unsw_nb15_icarl[-1])
```

```python
plt.plot(unsw_nb15_der, label="der")
plt.plot(unsw_nb15_finetune, label="finetune")
plt.plot(unsw_nb15_memo, label="memo")
plt.plot(unsw_nb15_lwf, label="lwf")
plt.plot(unsw_nb15_icarl, label="icarl")
plt.legend()
plt.xlabel("Task")
plt.ylabel("Accuracy")
plt.savefig("unsw_nb15_accy.png")
```
